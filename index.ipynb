```python
#Account Table Creation
account_table_sql = """
CREATE TABLE IF NOT EXISTS accounts (
    customer_id varchar(50)  PRIMARY KEY,
    account_balance DECIMAL(12,2),
    last_updated TIMESTAMP  
)
"""
connection = get_connection()
try:
    with connection.cursor() as cursor:
        cursor.execute(account_table_sql)
    connection.commit()
    print("Account Table created successfully")
finally:
    connection.close()
```

```python
# Q1: How many customers exist per city, and what is their average account balance?
query = """
SELECT 
c.city, 
count(  c.customer_id) as `Customer Count Per City`,
round(AVG(a.account_balance),2) AS `Average Account Balance`
FROM customers c
JOIN accounts a
    ON c.customer_id = a.customer_id
GROUP BY c.city
ORDER BY `Customer Count Per City` DESC
"""
connection = get_connection()
try:
    with connection.cursor() as cursor:
        cursor.execute(query)
        rows = cursor.fetchall()
        df = pd.DataFrame(rows)
finally:
    connection.close()
df
```

```python
#Q13: How many customers exist in each age group (e.g., 18â€“25, 26â€“35, etc.)?

query = """
select 
  CASE
    WHEN `age` BETWEEN 18 AND 25 THEN '18â€“25'
    WHEN `age` BETWEEN 26 AND 35 THEN '26â€“35'
    WHEN `age` BETWEEN 36 AND 45 THEN '36â€“45'
    WHEN `age` BETWEEN 46 AND 55 THEN '46â€“55'
    WHEN `age` BETWEEN 56 AND 65 THEN '56â€“65'
    ELSE '66+'
  END AS `age_group`,
count(*) as `Count of Each Age`
from banksight.customers
GROUP BY `age_group`
order by `age_group`

"""
connection = get_connection()
try:
    with connection.cursor() as cursor:
        cursor.execute(query)
        rows = cursor.fetchall()
        df = pd.DataFrame(rows)
finally:
    connection.close()
df
```

```python
#Q14: Which issue categories have the longest average resolution time?

query = """
SELECT 
    `Issue_Category`,
    AVG(DATEDIFF(`Date_Closed`, `Date_Opened`)) AS `Average Resolution Days`
FROM banksight.support_tickets
WHERE 
    `Date_Closed` IS NOT NULL
    AND `status` = 'Closed'
GROUP BY `Issue_Category`
HAVING AVG(DATEDIFF(`Date_Closed`, `Date_Opened`)) IS NOT NULL
ORDER BY `Average Resolution Days` DESC;
"""
connection = get_connection()
try:
    with connection.cursor() as cursor:
        cursor.execute(query)
        rows = cursor.fetchall()
        df = pd.DataFrame(rows)
finally:
    connection.close()
df
```
